# Görüntü Verisi

Görüntü verileri konum-renk bilgilerini tutan bir tiptir. Bilgisayarda bu verileri saklamanın bir çok yolu vardır fakat en çok tercih edilen method RGB formatındaki tensörlerdir. RGB, her bir piksel için kırmızı, yeşil ve mavi olmak üzere 3 ana rengin değerlerini 0-255 arasında tutar. Bir resimde gördüğünüz mavi pikselin değeri [0,0,255] olarak saklanır. Ana renklerin karışımından tüm renkler elde edilebilir ve bir ana rengi temsil eden değer 255'e ne kadar yakınsa o piksel için o ana renk o kadar yoğundur.

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
siyah = np.full((100,100),0)
```

```python
siyah = Image.fromarray(siyah)
```

```python
plt.imshow(siyah)
```

```python
beyaz = np.full((150,150),255)
beyaz = Image.fromarray(beyaz)
plt.imshow(beyaz)
```

```python
mavi = np.random.randint(255,size=(200,200))
kırmızı = np.random.randint(255,size=(200,200))
yesil = np.random.randint(255,size=(200,200))

resim = np.dstack((kırmızı,yesil,mavi)).astype(np.uint8)

resim = Image.fromarray(resim)
plt.imshow(resim)
```
